```python
import pandas as pd 
import numpy as np
import os
from collections import defaultdict
```

```python
folder = [r'small_task1_filt',r'small_task2_filt']
size = defaultdict(lambda: defaultdict(int))
for fol in folder:
    for file in os.listdir(fol):
        path = os.path.join(fol,file)
        lang = path.split(".")
        fil = lang[-1] == "filt"
        lang0 = lang[-2] if fil else lang[-1]
        lang1 = lang[-3].split("-") if fil else lang[-2].split("-")
        lang1.remove(lang0)
        try:
            with open(path, encoding="utf-8") as f:
                for i,l in enumerate(f):
                    pass
            length = i + 1
            print(path, '::', length, end=' '*50+'\r')
            size [lang0] [lang1[0]] += length
        except (UnicodeDecodeError) as e:
            print("error: ", path)
            pass
```

```python
df = pd.DataFrame.from_dict(size).loc[['en','et', 'hr', 'hu', 'sr', 'mk', 'id', 'tl', 'jv', 'ms', 'su', 'ta']]
df.fillna(0).astype(int)
```
